```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```
#Pega csvs de repositorios publico
import requests
url = 'https://raw.githubusercontent.com/laxmimerit/Building-Your-First-ANN-with-TensorFlow-2.0/master/Customer_Churn_Modelling.csv'
res = requests.get(url, allow_redirects=True)
with open('Customer_Churn_Modelling.csv','wb') as file:
    file.write(res.content)
df = pd.read_csv('Customer_Churn_Modelling.csv')
```

```
X = df.drop(["RowNumber","CustomerId","Surname","Exited"], axis=1)
y = df["Exited"]
```

```
from sklearn.preprocessing import StandardScaler
```
